```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/HarryPotterGradesDataset.csv'
df = pd.read_csv(file_path, header=0)
print("All Datas:")
print(df)

```

```python
# print("Dataset Shape:" ,df.shape)
# print("First 5 rows:")
# print(df.head())
# print(df.tail())

# print('Column Names:')
# print(df.columns.tolist())

# print("Missing Values per column:")
# print(df.isnull().sum())

# print("Percentage of missing values per column:")
# print((df.isnull().sum()/len(df))*100)

# mean_value=df['Potions'].mean()
# mean=df.iloc[1]
# print("Mean:" , mean_value)

median_value=df['Potions'].median()
median=df.iloc[1]
#print("Median:" , median_value)

# mode_value=df['Potions'].mode()
# mode=df.iloc[1]
# print("mode:" , mode_value)

# keep the original score
df["Potions_Original"]=df['Potions']

# fill the missing values mean
# Imputation - filling in missing values
df['Potion_filled']=df['Potions'].fillna(median_value) #imputation
# print(df)

print("Median value used for imputation" , median_value)
# print("Head:\n" , df.head())
# print("Tail:\n" , df.tail())

#plot the datase
plt.figure(figsize=(12,6))
plt.plot(df['Name'],df['Potions_Original'],'bo-',label='Original(with missing)')
plt.plot(df['Name'],df['Potion_filled'],'rx--',label=f'Imputed (Median={median_value:.2f})')
plt.xticks(rotation=45)
plt.title("Potions Grade: Original (Blue) vs Imputed (Red)")
plt.xlabel("Student Name")
plt.ylabel("Potion Grade")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ========== STEP 1/9: Import necessary libraries ==========
# para handling ug analyze data, tables
import pandas as pd
# para sa numbers
import numpy as np
from sklearn.impute import SimpleImputer
# para magamit ang drive
from google.colab import drive
drive.mount('/content/drive')

# ========== STEP 2/9: import the dataset ==========
file_path = '/content/drive/MyDrive/HarryPotterGrades Dataset.csv'
data3 = pd.read_csv(file_path)
print("All Datas:")
print(data3)

# ========== STEP 3/9: Load the dataset ==========
X = data3.iloc[:, :-1].values
y = data3.iloc[:, -1].values

print("\nDataset of X:")
print (X)

print("\nDataset of y:")
print(y)

# ========== STEP 4/9: Identify the missing data (NaN) ==========
print("\nAll Missing Datas:")
print(data3.isna())

# ========== STEP 5/9: Number of missing entries ==========
print("\nNumber of Missing Datas:")
print(data3.isna().sum())

# ========== STEP 6/9: Configure instance of SimpleImputer class ==========
# For X
imputer_x = SimpleImputer(missing_values=np.nan, strategy = 'mean')

# ========== STEP 7/9: Fit imputer on DataFrame ==========
# For X
imputer_x.fit(X[:, 1:])

# ========== STEP 8/9: Apply transform to DataFrame ==========
# For X
X[:, 1:] = imputer_x.transform(X[:, 1:])

# ========== STEP 9/9: Print updated matrix of features ==========
print("\nUpdated Matrix of Features:")
print(X)
```
